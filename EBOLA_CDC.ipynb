```python
import pandas as pd

# Replace 'Sheet1' with the actual sheet name if it's different
sheet_name = 'Sheet1'

# Provide the full path to the Excel file
file_path = r'C:\Users\sobin\Downloads\previous-case-counts.xlsx'

# Read the Excel file into a pandas dataframe
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Display the dataframe
df
```

```python
df['WHO report date'] = pd.to_datetime(df['WHO report date'])
# Set 'WHO report date' as the index of the dataframe
#df.set_index('WHO report date', inplace=True)

# Reverse the order of rows in the dataframe
df= df.sort_index(ascending=False)

# Reset the index without keeping it as a column
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
# Iterate through columns starting from the second one
for col in df.columns[1:]:
    for i in range(1, len(df)):
        # If the current value is less than the previous one, update it
        if df.loc[i, col] < df.loc[i - 1, col]:
            df.loc[i, col] = df.loc[i - 1, col]
```

```python
df
```

```python

# Calculate the daily new cases for Guinea
df['Cases in Guinea'] = df['Total Cases, Guinea'] - df['Total Cases, Guinea'].shift(1)
# Fill the NaN in the first row with the actual first value
df['Cases in Guinea'].fillna(df['Total Cases, Guinea'].iloc[0], inplace=True)
df
```

```python
# Calculate the daily new Deaths for Guinea
df['Deaths in Guinea'] = df['Total Deaths, Guinea'] - df['Total Deaths, Guinea'].shift(1)
# Fill the NaN in the first row with the actual first value
df['Deaths in Guinea'].fillna(df['Total Deaths, Guinea'].iloc[0], inplace=True)
df
```

```python
# Calculate the daily new cases for Liberia
df['Cases in Liberia'] = df['Total Cases, Liberia'] - df['Total Cases, Liberia'].shift(1)

# Fill the NaN in the first row with the actual first value
df['Cases in Liberia'].fillna(df['Total Cases, Liberia'].iloc[0], inplace=True)
```

```python
# Calculate the daily new deaths for Liberia
df['Deaths in Liberia'] = df['Total Deaths, Liberia'] - df['Total Deaths, Liberia'].shift(1)

# Fill the NaN in the first row with the actual first value
df['Deaths in Liberia'].fillna(df['Total Deaths, Liberia'].iloc[0], inplace=True)
```

```python
df
```

```python
# Calculate the daily new cases for Sierra Leone
df['Cases in Sierra Leone'] = df['Total Cases, Sierra Leone'] - df['Total Cases, Sierra Leone'].shift(1)

# Fill the NaN in the first row with the actual first value
df['Cases in Sierra Leone'].fillna(df['Total Cases, Sierra Leone'].iloc[0], inplace=True)

# Calculate the daily new deaths for Sierra Leone
df['Deaths in Sierra Leone'] = df['Total Deaths, Sierra Leone'] - df['Total Deaths, Sierra Leone'].shift(1)

# Fill the NaN in the first row with the actual first value
df['Deaths in Sierra Leone'].fillna(df['Total Deaths, Sierra Leone'].iloc[0], inplace=True)
```

```python
df
```

```python
# Drop columns from 2nd to 7th
df.drop(df.columns[1:9], axis=1, inplace=True)# Drop columns from 2nd to 7th
df
```

```python
# Fill negative values with zero
df = df.applymap(lambda x: x if (isinstance(x, pd.Timestamp) or x >= 0) else 0)
```

```python
df
```

```python
# Convert 'WHO report date' column to datetime format
df['WHO report date'] = pd.to_datetime(df['WHO report date'])
```

```python
# Extract week number and year from the 'WHO report date' column
df['Week'] = df['WHO report date'].dt.strftime('%Y-%U')
```

```python
# Drop the original 'WHO report date' column
df.drop('WHO report date', axis=1, inplace=True)
```

```python
df
```

```python
# Group by 'Week' and sum the values for each country
df = (
    df.groupby('Week')
    .agg({
        'Cases in Guinea': 'sum',
        'Deaths in Guinea': 'sum',
        'Cases in Liberia': 'sum',
        'Deaths in Liberia': 'sum',
        'Cases in Sierra Leone': 'sum',
        'Deaths in Sierra Leone': 'sum'
    })
    .reset_index()
)
```

```python
df
```

```python
# Create a DataFrame with a continuous range of weeks in year-week format
all_weeks = pd.DataFrame({'Week': pd.date_range(start='2014-03-24', end='2016-04-13', freq='W-Mon')})

# Convert 'Week' column in all_weeks to year-week format
all_weeks['Week'] = all_weeks['Week'].dt.strftime('%Y-%U')

# Merge the continuous range of weeks with the original DataFrame
grouped_df = (
    all_weeks
    .merge(df, on='Week', how='left')
    .fillna(0)
)
```
